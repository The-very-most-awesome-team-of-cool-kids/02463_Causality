# Statistics

```python
print("_"*20, "\033[1mNo interventions\033[0m", "_"*20)
print("\033[1mCorrelations:\033[0m\n", data_none.corr())
print("\n")
print("\033[1mMeans:\033[0m\n", data_none.mean(axis = 0))
print("\n")
print("\033[1mCovariance:\033[0m\n", data_none.cov())
print("_"*50)

print("_"*20, "\033[1mS interventions\033[0m", "_"*20)
print("\033[1mCorrelations:\033[0m\n", data_s1.corr())
print("\n")
print("\033[1mMeans:\033[0m\n", data_s1.mean(axis = 0))
print("\n")
print("\033[1mCovariance:\033[0m\n", data_s1.cov())
print("_"*50)

print("_"*20, "\033[1mQ interventions\033[0m", "_"*20)
print("\033[1mCorrelations:\033[0m\n", data_q1.corr())
print("\n")
print("\033[1mMeans:\033[0m\n", data_q1.mean(axis = 0))
print("\n")
print("\033[1mCovariance:\033[0m\n", data_q1.cov())
print("_"*50)

print("_"*20, "\033[1mR interventions\033[0m", "_"*20)
print("\033[1mCorrelations:\033[0m\n", data_r1.corr())
print("\n")
print("\033[1mMeans:\033[0m\n", data_r1.mean(axis = 0))
print("\n")
print("\033[1mCovariance:\033[0m\n", data_r1.cov())
print("_"*50)

print("_"*20, "\033[1mT interventions\033[0m", "_"*20)
print("\033[1mCorrelations:\033[0m\n", data_t1.corr())
print("\n")
print("\033[1mMeans:\033[0m\n", data_t1.mean(axis = 0))
print("\n")
print("\033[1mCovariance:\033[0m\n", data_t1.cov())
print("_"*50)

print("_"*20, "\033[1mH interventions\033[0m", "_"*20)
print("\033[1mCorrelations:\033[0m\n", data_h1.corr())
print("\n")
print("\033[1mMeans:\033[0m\n", data_h1.mean(axis = 0))
print("\n")
print("\033[1mCovariance:\033[0m\n", data_h1.cov())
print("_"*50)
```

```python
no_means = data_none.mean()
q1_means = data_q1.mean()
r1_means = data_r1.mean()
s1_means = data_s1.mean()
t1_means = data_t1.mean()
h1_means = data_h1.mean()

means = pd.DataFrame(list(zip(no_means, q1_means, r1_means, s1_means, t1_means, h1_means)), 
             columns = ["None", "Q1", "R1", "S1", "T1", "H1"], index = ["Q", "R", "S", "T"])


print("\033[1mMeans with the different interventions\033[0m")
display(means)

mean_diff = means.sub(means["None"], axis = 0).divide(means["None"], axis = 0)
diff_bound = .5

print("\n\033[1mThe difference in means when making interventions\033[0m")
mean_diff = mean_diff.style.applymap(lambda x: 'background-color : yellow' if abs(x)>diff_bound else '')
display(mean_diff)

no_var = data_none.var()
q1_var = data_q1.var()
r1_var = data_r1.var()
s1_var = data_s1.var()
t1_var = data_t1.var()
h1_var = data_h1.var()


var = pd.DataFrame(list(zip(no_var, q1_var, r1_var, s1_var, t1_var, h1_var)), 
             columns = ["None", "Q1", "R1", "S1", "T1", "H1"], index = ["Q", "R", "S", "T"])

print("\033[1mVariance with the different interventions\033[0m")
display(var)

var_diff = var.sub(var["None"], axis = 0).divide(var["None"], axis = 0)
diff_bound = .5

print("\n\033[1mThe difference in variance when making interventions\033[0m")
var_diff = var_diff.style.applymap(lambda x: 'background-color : yellow' if abs(x)>diff_bound else '')
display(var_diff)
```
